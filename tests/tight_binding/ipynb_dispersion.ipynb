# Honeycomb Tight-Binding Model

Let's setup the problem class using the honeycomb data json.

```python
from acceptor_TI import Problem

data_path = "../../../acceptor_TI/data/"
file_name = "honeycomb.json"

problem = Problem(data_path=data_path, file_name=file_name)

problem.setup(
    size = 10, # size of lattice in real space
    N_k = 300, # reciprocal space discretization
    dispersion = True # solve eigenvalue problem
)
```

```python
problem.plot(type="lattice")
```

```python
problem.run()
```

```python
problem.plot("dispersion")
```

We can modify this data without hardcoding values. e.g. random.

```python
hopping_dict = problem.cell_parser.eigenvalues.nn_hopping.value

hopping_dict["t_ss_sigma"] = -1.4
hopping_dict["t_sp_sigma"] = 0.2
hopping_dict["t_pp_sigma"] = 0.4
hopping_dict["t_pp_pi"] = -0.81

print(problem.cell_parser.eigenvalues.nn_hopping.value)
```

```python
problem.setup(
    size = 10,
    N_k = 500, 
    dispersion = True
)

problem.run()
problem.plot("dispersion")
```
